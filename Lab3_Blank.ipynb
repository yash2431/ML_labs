### 3.  Train_Test_Split
**Concept:** We split data to prevent "Overfitting". The model learns from the **Train** set and is tested on the **Test** set.

**Task:** 
1. Define `X` (Features: everything except Rating/App) and `y` (Target: Rating).
2. Split the data (80% Train, 20% Test).

```python
# train_test_split
from sklearn.model_selection import train_test_split

x = df.drop(['Rating','App'],axis=1)
y = df['Rating']

# Split the data into 80% training and 20% testing sets
x_train, x_test, y_train, y_test = train_test_split(
    x, y, 
    test_size=0.2, 
    random_state=42, # You can use any integer for reproducibility
    shuffle=True # Shuffling is True by default, which is generally good practice
)

# Print the shapes of the resulting sets to verify the split
print(f'X_train shape: {x_train.shape}')
print(f'X_test shape: {x_test.shape}')
print(f'y_train shape: {y_train.shape}')
print(f'y_test shape: {y_test.shape}')
```

### 4. 📏 Scaling Numerical Data (StandardScaler)
**Concept:** `Installs` (Millions) are much larger than `Rating` (1-5). We scale them so the model treats them equally.

**Task:** Use `StandardScaler` on the numerical columns.

```python
#StandardScaler()
#fit_transform()

from sklearn.preprocessing import StandardScaler

# Select only numerical columns (e.g., Installs, Rating, Reviews, Size, Price)
# select_dtypes helps us automatically filter numeric columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Create a StandardScaler object
# StandardScaler transforms features so they have mean = 0 and standard deviation = 1
scaler = StandardScaler()

# Make a copy of the original dataframe to store scaled values
df_scaled = df.copy()

# Apply scaling only to numerical columns
# fit_transform learns the scaling parameters AND applies them
df_scaled[num_cols] = scaler.fit_transform(df[num_cols])

# Display first 5 rows of the scaled data
df_scaled.head()
```

### 5. 🔠 Encoding Categorical Data
**Concept:** Models need numbers, not text like "Business" or "Teen".

**Method A: Pandas `get_dummies` (Simple)**

```python
#get_dummies

# Select categorical columns to encode
# These are columns with text (object type)
cat_cols = df.select_dtypes(include=['object']).columns.tolist()

# Apply one-hot encoding using get_dummies()
# drop_first=True removes one dummy column per category to avoid multicollinearity (optional)
df_encoded = pd.get_dummies(df, columns=cat_cols, drop_first=True)

# Show first 5 rows of encoded dataset
df_encoded.head()
```

**Method B: Sklearn `OneHotEncoder` (Professional)**

```python
#OneHotEncoder
#fit_transform
from sklearn.preprocessing import OneHotEncoder

# Machine learning models cannot understand text categories like:
# "Business", "Tools", "Teen", "Everyone", etc.
# So we extract only columns containing text (object dtype)
cat_cols = df.select_dtypes(include=['object']).columns.tolist()

# -----------------------------
# 2. Create OneHotEncoder object
# -----------------------------
# handle_unknown='ignore' helps when new unseen categories appear later
# sparse=False returns a normal NumPy array instead of sparse matrix
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

# -----------------------------
# 3. Fit and transform the categorical columns
# -----------------------------
# fit_transform learns categories AND converts them to numeric dummy columns
encoded_array = encoder.fit_transform(df[cat_cols])

# -----------------------------
# 4. Convert encoded output into a DataFrame
# -----------------------------
# get_feature_names_out gives clean column names like "Category_Tools"
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(cat_cols))

# -----------------------------
# 5. Concatenate with original dataset (drop old text columns)
# -----------------------------
df_encoded = pd.concat([df.drop(columns=cat_cols), encoded_df], axis=1)

# Display first 5 rows
df_encoded.head()
```

### 6. 🚀 The Full Pipeline: ColumnTransformer
**Concept:** Instead of doing steps 4 and 5 manually, we wrap them in one object.

**Task:** Create a `ColumnTransformer` that Scales numerical data AND Encodes categorical data at the same time.

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
# Create ColumnTransformer

# ------------------------------------------
# 1. Identify Numerical and Categorical Columns
# ------------------------------------------

# Numerical = int and float columns (e.g., Installs, Rating, Reviews, Size)
num_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Categorical = object (text) columns (e.g., Category, Type, Content Rating)
cat_cols = df.select_dtypes(include=['object']).columns.tolist()

# ------------------------------------------
# 2. Create ColumnTransformer
# ------------------------------------------
# This object performs BOTH scaling + encoding together

preprocessor = ColumnTransformer(
    transformers=[
        # Scale all numeric columns
        ('num', StandardScaler(), num_cols),

        # Encode all categorical columns
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), cat_cols)
    ]
)

# ------------------------------------------
# 3. Apply the transformations
# ------------------------------------------
# fit_transform() learns from the data AND applies transformations

processed_data = preprocessor.fit_transform(df)

# ------------------------------------------
# 4. Convert to DataFrame (optional but helpful)
# ------------------------------------------

# Get encoded categorical feature names
cat_features = preprocessor.named_transformers_['cat'].get_feature_names_out(cat_cols)

# Combine numeric + categorical column names
final_columns = list(num_cols) + list(cat_features)

# Build the final DataFrame
df_processed = pd.DataFrame(processed_data, columns=final_columns)

# Show first 5 rows
df_processed.head()
```

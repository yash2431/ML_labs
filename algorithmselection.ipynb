## Step-1: Import Required Libraries

```python
import numpy as np
import pandas as pd
```

## Step-2: Load the Dataset

```python
# Read CSV file and store data in DataFrame
data = pd.read_csv("insurance_fraud_data_cleaned.csv")

# Display first 5 rows to understand dataset structure
data.head()
```

## Step-3: Split Input (X) and Output (y)

```python
# iloc selects rows and columns using index positions

# X contains all feature columns (independent variables)
X = data.iloc[:, :-1].values    

# y contains the target column (dependent variable: Fraud / Not Fraud)
y = data.iloc[:, -1].values     

# Reshape y into column vector for matrix compatibility
y = y.reshape(-1, 1)
```

## Step-4: Feature Scaling (Standardization)

```python
# Scaling improves convergence speed of gradient descent
# Formula: (value - mean) / standard deviation

X = (X - X.mean(axis=0)) / X.std(axis=0)
```

## Step-5: Initialize Model Parameters

```python
# m → number of training examples (rows)
# n → number of features (columns)
m, n = X.shape

# Initialize weights with zeros (one weight per feature)
weights = np.zeros((n, 1))

# Initialize bias to zero
bias = 0

# Learning rate controls step size of gradient descent
learning_rate = 0.01

# Number of iterations to train the model
epochs = 1000
```

## Step-6: Define Sigmoid Function

```python
def sigmoid(z):
    """
    Sigmoid converts linear output into probability between 0 and 1
    Formula: 1 / (1 + e^-z)
    """
    return 1 / (1 + np.exp(-z))
```

## Step-7: Train Model using Gradient Descent

```python
for i in range(epochs):

    # Step 1: Linear combination (z = Xw + b)
    z = np.dot(X, weights) + bias

    # Step 2: Apply sigmoid to get predicted probability
    y_pred = sigmoid(z)

    # Step 3: Compute Binary Cross Entropy loss
    # E(y,y') = -[ylog(y')+(1-y)log(1-y')]
    # Binary Cross-Entropy is a loss function used in binary classification problems, where the target variable has only two classes:
    # 0 → Negative class (Not Fraud)
    # 1 → Positive class (Fraud)
    # It measures how far the predicted probability is from the actual label.
    cost = -np.mean(
        y * np.log(y_pred + 1e-9) + 
        (1 - y) * np.log(1 - y_pred + 1e-9)
    )

    # Step 4: Calculate gradients
    # dw → gradient of cost w.r.t weights
    dw = (1 / m) * np.dot(X.T, (y_pred - y))

    # db → gradient of cost w.r.t bias
    db = (1 / m) * np.sum(y_pred - y)

    # Step 5: Update weights and bias
    weights -= learning_rate * dw
    bias -= learning_rate * db

    # Print cost every 100 epochs to monitor learning
    if i % 100 == 0:
        print(f"Epoch {i}, Cost: {cost}")
```

## Step-8: Prediction Function

```python
def predict(X, weights, bias):
    """
    Predicts class labels (0 or 1)
    1 → Fraud
    0 → Not Fraud
    """
    z = np.dot(X, weights) + bias
    probabilities = sigmoid(z)

    # Convert probabilities to binary output using threshold 0.5
    return (probabilities >= 0.5).astype(int)
```

## Step-9: Model Evaluation (Accuracy)

```python
# Get predictions for training data
y_pred = predict(X, weights, bias)

# Accuracy = Correct predictions / Total predictions
accuracy = np.mean(y_pred == y) * 100

print("Model Accuracy:", accuracy, "%")
```

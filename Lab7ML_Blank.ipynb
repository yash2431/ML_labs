# Decision Tree Classifier

```python
from sklearn.datasets import load_iris 
```

# Import iris data ser using sklearn

```python
data = load_iris()
```

```python
data
```

# Importing the libraries

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import tree
import matplotlib.pyplot as plt
```

# Divide the data into input and output

```python
x = data['data']
y = data['target']
```

```python
x
```

```python
y
```

# Splitting the dataset into the Training set and Test set

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Fitting DecisionTreeClassifier   on dataset

```python
model = DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

# Display Decision Tree

```python
%matplotlib inline
```

```python
tree.plot_tree(model)
```

```python
plt.show()
```

# Predict the x_test 

```python
y_predict = model.predict(x_test)
```

```python
y_predict
```

```python
y_test
```

```python
model.score(x_train,y_train)
```

# Import diabetes.csv dataset 

```python
df = pd.read_csv('diabetes.csv')
```

```python
df
```

# Check the distribution of the target

```python
df['Outcome'].value_counts()
```

# Replace zeros with NaN for specific columns

```python
cols_with_zero = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
df[cols_with_zero] = df[cols_with_zero].replace(0, np.nan)
```

# Check for missing values

```python
df.isnull().sum()
```

# Fill missing values with median

```python
for col in cols_with_zero:
    df[col].fillna(df[col].median(),inplace=True)
```

```python
df[cols_with_zero].isnull().sum()
```

```python
df.isnull().sum()
```

# Visualize Distributions

```python
import matplotlib.pyplot as plt

for col in cols_with_zero:
    plt.figure()
    plt.hist(df[col], bins=30)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

# Convert Targer data into interger code

```python
df['Outcome'] = df['Outcome'].map({'Diabetic':1,'Non Diabetic':0})
```

```python
df
```

# SVR

# Importing the libraries

```python
import numpy as np
import pandas as pd
```

# Read World bank CSV

```python
df = pd.read_csv("WorldBank.csv")
```

```python
df
```

# Perform conditional selection to find - Population ages 15-64 (% of total population)

```python
population_15_64 = df[df["Indicator Name"] == "Population ages 15-64 (% of total population)"]
```

```python
population_15_64
```

# Divide the data into input and output

```python
x = np.arange(1960,2020).reshape(-1,1)
```

```python
x
```

```python
y = population_15_64.values[0][4:-1].reshape(-1,1)
```

```python
y
```

# Plot scatter plot of Population ages 15-64 (% of total population)

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.show()
```

# Feature Scaling (Mandatory for SVR)**
SVR is highly sensitive to the range of data points. If we don't scale (normalize) the data, the model will fail to find the correct hyperplane.

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc_x = StandardScaler()
sc_y = StandardScaler()

x_scaled = sc_x.fit_transform(x)
y_scaled = sc_y.fit_transform(y)

print("Scaling complete")
```

# Fitting SVR on Linear Kernel on dataset

```python
from sklearn.svm import SVR
```

```python
svr = SVR(kernel='linear')
svr.fit(x_scaled,y_scaled.ravel())
```

# Predict Values using Model

```python
# use inverse_transform
```

```python
y_pred_scaled = svr.predict(x_scaled)
y_pred = sc_y.inverse_transform(y_pred_scaled.reshape(-1, 1))
```

```python
plt.scatter(x,y)
plt.scatter(x,y_pred)
plt.show()
```

# Predict the X_scaled using 3 Kernel

```python
# above Steps Only - Kernel Change
```

```python
svr_lin = SVR(kernel='linear')
svr_lin.fit(x_scaled,y_scaled.ravel())

svr_poly = SVR(kernel='poly',degree=3)
svr_poly.fit(x_scaled,y_scaled.ravel())

svr_rbf = SVR(kernel='rbf')
svr_rbf.fit(x_scaled,y_scaled.ravel())
```

```python
y_pred_scaled_lin = svr_lin.predict(x_scaled)
y_pred_lin = sc_y.inverse_transform(y_pred_scaled_lin.reshape(-1, 1))

y_pred_scaled_poly = svr_poly.predict(x_scaled)
y_pred_poly = sc_y.inverse_transform(y_pred_scaled_poly.reshape(-1, 1))

y_pred_scaled_rbf = svr_rbf.predict(x_scaled)
y_pred_rbf = sc_y.inverse_transform(y_pred_scaled_rbf.reshape(-1, 1))
```

# Visualising the  results

```python
plt.scatter(x, y, label="Actual Data")
plt.plot(x, y_pred_lin, label="Linear Model")
plt.plot(x, y_pred_poly, label="Polynomial Model")
plt.plot(x, y_pred_rbf, label="RBF Model")
plt.legend()
plt.show()
```

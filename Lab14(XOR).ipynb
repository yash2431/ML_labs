```python
def step_function(x):
    if x>=0:
        return 1
    else:
        return 0

print('Activation function done')
```

```python
import numpy as np

x = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y = np.array([0,1,1,0])
```

```python
weights = np.array([0,0])
b = 0
learning_rate = 0.5
```

```python
input_sample = x[1]
target_value = y[1]

z = np.dot(input_sample,weights) + b
print(f'Math: ({input_sample[0]} * {weights[0]}) + ({input_sample[1]} * {weights[1]}) + {b}')
print(f'Sum (z): {z:.2f}')

pred = step_function(z)
print(f'Prediction: {pred}')

error = target_value - pred
print(f'Error: {error}')
```

```python
epochs = 5
print('---------Training-------------')

for epoch in range(epochs):
    print(f'\nEPOCH: {epoch + 1}')

    for i in range(4):
        x_or = x[i]
        y_or = y[i]

        z = np.dot(x_or, weights) + b
        prediction = step_function(z)

        error = y_or - prediction

        if error != 0:
            weights = weights + (learning_rate * error * x_or)
            b = b + (learning_rate * error * 1)
            print(f'MISTAKE on {x_or}: Expected {y_or} got {prediction}')
            print(f'Updated Weights: {weights} , Bias: {b:.2f}')
        else:
            print(f'Correct on {x_or}')

print('\nFinal Weights:',weights)
print('Final Bias', b)
```
